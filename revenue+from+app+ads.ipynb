## part 2: we wrote a function that detects non-English app names

### in this code:

-we make a new function that takes in the string and returns "True if the string in english language  or has at maximum 3 (emojis or other symbols) " or "False if the string in another langauge or has more than 3 (emojis or other symbols) ".
    -we use loop in the string or "app name".
    
     - then use if to check the ord() of the string is more than 127 or less than it.
    -if the ord() of the string is higher than 127 it return "False". and if the ord() of the string is lower than or equal 127 it return "True"

```python
def App_english(app_name):
    non_english=0
    for char in app_name:
        if ord(char)  > 127 :
            non_english+=1
            
    if non_english > 3:
        return False
    else:
         return True 
```

```python
#free_android=[]
#free_IOS=[]

def freq_table(data_clean, index):
    table={}
    total=0
    
    for row in data_clean:
        total+=1
        value = row[index] 
        if value in table:
            table[value]+=1
        else:
            table[value]=1
            
            
    table_precentages={}
    
    for key in table:
        precentage=(table[key]/total)*100
        table_precentages[key]=precentage
    return table_precentages
            
            
        
def display_table(data_clean, index):
    table = freq_table(data_clean, index)
    display = []
    for key in table:
        key_value=(table[key],key)
        display.append(key_value)
        
    table_stored= sorted(display,reverse = True)
    for row in table_stored:
        print(row[1], ':', row[0])
        
```
